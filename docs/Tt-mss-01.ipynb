Through a lot of trial and error, we learned that some talks have only a few words but other talks have no words at all and are stored not as strings at all so they cause errors when trying to coundt words. I finally arrived at this odd bit of code that tests to see if the content of an item in `talks` is a string and if it's not, to replace it with an empty string.

```python
# We are going to load the texts/talks from the first set, and I'm going to make
# sure to remember which is which through these clunky variable names.

talks_1 = df1.text.tolist()

def string_test(s):
    if s is None:
        return ''
    else:
        return str(s)

# With any luck, this list comprehension will work:
strings_1 = [ string_test(text) for text in talks_1 ]
```

With all our talks cleaned into strings, we can now count the words in each talk. After that, we are going to create a separate, sorted list so we can see for ourselves the low and high numbers. We are keeping the original unsorted list so that we can use it to identify the talks we want to delete.

```python
import re

word_counts_1 = []
for talk in strings_1:
    count = len(re.findall(r"[a-zA-Z_]+", talk))
    word_counts_1.append(count)
    
print(sorted(word_counts_1))
```
