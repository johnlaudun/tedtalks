The data being used here is the filtered CSV of just the pronouns in the subject position.

```python
import pandas as pd

svo_data = pd.read_csv('../output/svos-pronouns.csv', index_col = 0)
# Index col: https://www.statology.org/pandas-use-first-column-as-index/
```

```python
test = svo_data[svo_data["subject"] == "he"]
```

```python
test2 = test.copy()
```

```python
test2.groupby(["verb"]).count()
```

```python
grp = test2.groupby("verb")
```

```python
for verb, group in grp:
    print(verb)
    print(len(group))
    print()
```

```python
verb_counts = grp.size().reset_index(name='obs').sort_values(['obs'], ascending=False)

#https://www.statology.org/pandas-groupby-count/
```

```python
# Drop all the verbs used just once

inds = verb_counts["obs"] > 1
```
